```python
def open_json(filename):
    with open(filename, "r", encoding="utf-8") as read_file:
        data = json.load(read_file)
    return data
```

```python
def add_children(tree):
    result = tree
    for i in tree:
        if "mid" in i.keys():
            if "cids" in result[i["mid"] - 1]:
                result[i["mid"] - 1]["cids"].append(i["id"])
            else:
                result[i["mid"] - 1].update({"cids":[i["id"]]})
        if "fid" in i.keys():
            if "cids" in result[i["fid"] - 1]:
                result[i["fid"] - 1]["cids"].append(i["id"])
            else:
                result[i["fid"] - 1].update({"cids":[i["id"]]})
    return result
```

```python
add_children(open_json('data/tree3.json'))
```

```python
def add_rank(tree, root_id):
    result = tree
    
    def add_rank_node(node, rank):
        if "rank" not in node.keys():
            node.update({"rank":str(rank)})
            if "pids" in node.keys():
                for n in node["pids"]:
                    add_rank_node(result[n-1], rank)
        else:
            rank+=1
            if "cids" in node.keys():
                for n in node["cids"]:
                    add_rank_node(result[n-1], rank)
    
    add_rank_node(result[root_id - 1], 0)
    
    return result
```

```python
tree_rank = add_rank(add_children(open_json('data/tree3.json')), 1)
tree_rank
```

```python
#удалить
def list_to_dict(lst, center):
    result = {}
    if len(lst)%2 == 0:
        half = len(lst)/2
        ind = 0
        for i in np.arange(-half + 0.5, half - 0.4, 1.0):
            result.update({i + center:lst[ind]})
            ind+=1
    else:
        ind = 0
        for i in range(len(lst)//2 - len(lst) + 1, len(lst)//2 + 1):
            result.update({i+ center:lst[ind]})
            ind+=1
    return(result)
```

```python
#преобразоывавание к формату сигмы
def to_sigma_tree(tree):
    result = tree
    for node in result:
        cur_x = node['x']
        node.pop('x')
        node.update({'label':node['name']})
        node.pop('name')
        node.update({'x':cur_x})
        node.update({'y':int(node['rank'])})
        node.pop('rank')
        node.update({'size':10 - node['y']})
        node['id'] = str(node['id'])
        if 'pids' in node.keys():
            node.pop('pids')
        if 'cids' in node.keys():
            node.pop('cids')
        if 'mid' in node.keys():
            node.pop('mid')
        if 'fid' in node.keys():
            node.pop('fid')
        node.pop('gender')
    return result
```

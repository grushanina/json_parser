```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
fig = plt.figure()
G = nx.path_graph(4)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)
```

```python
fig = plt.figure()
tree = nx.random_tree(n=50, seed=0)
nx.draw(tree)
```

```python
fig = plt.figure()
pos = nx.spring_layout(tree)
nx.draw(tree, pos)
```

```python
def draw_random_tree(n):
    fig = plt.figure()
    tree = nx.random_tree(n=n, seed=0)
    nx.draw(tree, node_size=[1000/n]*n)
    fig = plt.figure()
    pos = nx.spring_layout(tree)
    nx.draw(tree, pos, node_size=[1000/n]*n)
```

```python
draw_random_tree(10)
```

```python
draw_random_tree(20)
```

```python
draw_random_tree(50)
```

```python
draw_random_tree(100)
```
